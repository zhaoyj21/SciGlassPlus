```python
##print para name
count = 0
for x in dois:
    if doc_dict[x]['content']:
        if len(doc_dict[x]['content'])>1:
            print(doc_dict[x]['content'][1]['sec_title'])
            count+=1
        else:
            print(x)
```

```python
#####example of error parse
doc_dict['10.1016/S1003-6326(19)65193-9']['content']
```

```python
##identify non table of rawdata
for x in dois:
    if len(doc_dict[x]['table']) == 0:
        print(x)
```

```python
#parse para-table pair
doc_dict2={}
tables = []
for x in dois:
    doc={}
    #####exist parse
    if doc_dict[x]['content']:
        #####not error parse
        if len(doc_dict[x]['content'])>1:
            #####screen para using section name
            if doc_dict[x]['content'][1]['sec_title'].count('Experiment') + doc_dict[x]['content'][1]['sec_title'].count('Method') + doc_dict[x]['content'][1]['sec_title'].count('Methodology')+ doc_dict[x]['content'][1]['sec_title'].count('method') + doc_dict[x]['content'][1]['sec_title'].count('Material') > 0:
                for i in range(len(doc_dict[x]['content'][1]['content'])):
                    tmp = []
                    if type(doc_dict[x]['content'][1]['content'][i]) == str:
                        tmp.append(str(doc_dict[x]['content'][1]['content'][i]))
                        break
                    if type(doc_dict[x]['content'][1]['content'][i]) == dict:
                        tmp.append(str(doc_dict[x]['content'][1]['content'][i]['content']))
                        break
                    if type(doc_dict[x]['content'][1]['content'][i]) == list:
                        if doc_dict[x]['content'][1]['content'][i][0]['content']:
                            tmp.append(str(doc_dict[x]['content'][1]['content'][i][0]['content'][0]))
                            break
                        else:
                            tmp.append(str(doc_dict[x]['content'][1]['content'][i][1]['content'][0]))
                            break
                #####exist table
                if doc_dict[x]['table']:
                    tmp5=[]
                    for j in range(len(doc_dict[x]['table'])):
                        tmp4 = tmp.copy()
                        tmp2 = json.dumps(doc_dict[x]['table'][j])
                        tmp3 = str(tmp2)
                        tables.append(tmp3)
                        tmp4.append(tmp3)
                        tmp5.append(tmp4)
                    doc['doi']=x
                    doc['content'] = tmp5
                    doc_dict2[x]=doc
                else:
                    print(x)
```
