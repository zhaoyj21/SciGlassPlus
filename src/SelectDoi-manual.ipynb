```python
#####identify different publishers
Elsevier = []
ASME = []
Emerald = []
IOP = []
MDPI = []
Nature = []
Sage = []
SPIE = []
Springer = []
TaylorFrancis = []
Wiley = []
doisel = [x.split('/') for x in select_doi_without_empty]
count = 0
for i in range(len(select_doi_without_empty)):
    if doisel[i][0] == '10.1016':
        Elsevier.append(select_doi_without_empty[i])
    elif doisel[i][0] == '10.1115':
        ASME.append(select_doi_without_empty[i])
    elif doisel[i][0] == '10.1108':
        Emerald.append(select_doi_without_empty[i])
    elif doisel[i][0] == '10.1088':
        IOP.append(select_doi_without_empty[i])
    elif doisel[i][0] == '10.3390':
        MDPI.append(select_doi_without_empty[i])
    elif doisel[i][0] == '10.1038':
        Nature.append(select_doi_without_empty[i])
    elif doisel[i][0] == '10.1177':
        Sage.append(select_doi_without_empty[i])
    elif doisel[i][0] == '10.1117':
        SPIE.append(select_doi_without_empty[i])
    elif doisel[i][0] == '10.1007' or doisel[i][0] == '10.1557':
        Springer.append(select_doi_without_empty[i])
    elif doisel[i][0] == '10.1080':
        TaylorFrancis.append(select_doi_without_empty[i])
    elif doisel[i][0] == '10.1002' or doisel[i][0] == '10.1111':
        Wiley.append(select_doi_without_empty[i])
    else:
        count+=1
        print(select_doi_without_empty[i])
```
